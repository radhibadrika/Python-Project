```python
import pandas as pd 
import seaborn as sns 
import matplotlib 
import numpy as np
import matplotlib.pyplot as plt 
plt.style.use ('ggplot')
from matplotlib.pyplot import figure 

%matplotlib inline 
matplotlib.rcParams ['figure.figsize'] = (12,8)

df = pd.read_csv (r"C:\Users\laksm\Documents\DATA ANALYST PROJECT\PHYTON\movies.csv")
```

```python
df.head ()
```

```python
#1 finding missing data 
# using for loop 

for col in df.columns: 
    pct_missing = np.mean(df[col].isnull())
    print('{} - {:.2%}'.format(col, pct_missing))
```

```python
#cleaning the data
df.cleaned = df.dropna()
```

```python
#checking if it was cleaned

df.cleaned.isnull ().sum ()
```

```python
df.dtypes
```

```python
#changing data types to remove ".0"

df['budget'] = df['budget'].fillna(0).astype('int64')
df['gross'] = df['gross'].fillna(0).astype('int64')
df['votes'] = df['votes'].fillna(0).astype('int64')
df['runtime'] = df['runtime'].fillna(0).astype('int64')
```

```python
df
```

```python
#Create correct year column

df['Correct_Year'] = df['released'].str.extract(r'(\d{4})')
df
```

```python
df = df.sort_values (by=['gross'], inplace = False , ascending = False)
```

```python
df
```

```python
# make scatterplot for budget vs gross 

plt.scatter (x=df['budget'], y=df ['gross'])
plt.title ('Budget vs Gross Earnings')
plt.ylabel ('Budget for Film')
plt.xlabel ('Gross Earnings')

plt.show ()
```

```python
df.head ()
```

```python
# plot budget vs gross using seaborn 

sns.regplot (x='budget', y='gross', data = df, scatter_kws = {"color":"red"}, line_kws = {"color":"blue"})
```

```python
#looking at correalation
#Using the pearson correlation 
#Results: "There's a high correlation between budget and gross in accordance to pearson correlation (0.5 - 0.7)"
#it appears that there's a high correlation between votes and gross 

df.corr (method = "pearson")
```

```python
#make visualizaiton for correlation matrix
# The brighter the color the higher the correlation 

correlation_matrix = df.corr (method = "pearson")
sns.heatmap (correlation_matrix, annot = True)
plt.title ('Correlation Metric for Numeric Features')
plt.ylabel ('Movie Features')
plt.xlabel ('Movie Features')
```

```python
#looking at what movie has the highest gross revenue per budget 
#it appears that "Paranormal Activity" film has the highest ROI 
# Side Note: Some of the budget are missing in this data, so we only include that have the data

df['ROI'] = df.apply(lambda row: row['gross'] / row['budget'] if row['budget'] > 0 else 0, axis=1)
max_roi_movie = df.loc[df['ROI'].idxmax()]
print("\nMovie with the highest ROI:")
print(max_roi_movie)






```
